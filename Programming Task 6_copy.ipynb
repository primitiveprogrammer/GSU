4. (10%) the range of y-axis is [-4, 4]
5. (10%) color: use red when the function is increasing and use green when it is decreasing.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 11, 1000)
x_values = np.arange(-5 * np.pi / 2, (7 * np.pi) / 2, 0.1)
ysin = 2 * np.sin(x)
ycos = 2 * np.cos(x)
```

```python
x_values = x[102:]
ysin_values = ysin[102:]
x2_values = x[27:925]
ycos_values = ycos[27:925]
```

```python
filter_ysin = np.cos(x_values) > 0
plt.plot(x_values, np.where(filter_ysin, ysin_values, np.nan), label = '$2sin(x)$', color = 'crimson')
plt.plot(x_values, np.where(~filter_ysin, ysin_values, np.nan), color = 'darkgreen')
filter_ycos = -np.sin(x2_values) > 0
plt.plot(x2_values, np.where(filter_ycos, ycos_values, np.nan), label = '$2cos(x)$', ls = '--', color = 'crimson')
plt.plot(x2_values, np.where(~filter_ycos, ycos_values, np.nan), ls = '--', color = 'darkgreen')
plt.yticks(np.arange(-4, 5))
plt.grid(color = 'b', alpha = 0.2)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.legend()
plt.show()
```

Q2. Consider following code:

    X = np.random.rand(1000) * 100
    Y = np.random.normal(50, 10, 1000)

1. (10%) Explain the meaning of the code. You may refer to the numpy documentation.
2. (10%) Calculate the mean and standard deviation of *X* and *Y*.
3. (10%) Plot the histograms of *X* and *Y*. Each histrogram should have 20 bins. Set appropriate alpha to make the overlapped area visible.
4. (10%) Describe the shape of each histogram. 

```python
X = np.random.rand(1000) * 100
Y = np.random.normal(50, 10, 1000)
```

random.rand creates an array and fills it with 1000 random numbers from 0-100. 
random.normal fills an array with random numbers from a distribution, so here it generates 100 numbers from 0-100 with a standard deviation of 10. 

```python
meanX = np.mean(X)
meanY = np.mean(Y)
print(meanX)
print(meanY)
```

```python
plt.hist(X, bins = 20, alpha = .5)
plt.hist(Y, bins = 20, alpha = .5)
plt.grid(alpha = 0.7)
plt.legend(['X', 'Y'])
```

The histogram for X is mostly flat, which i think indicates that the numbers are fairly evenly distributed. 
The histogram for Y is more of a pyramid shape, with the occurances of each number increasing as its value approaches 50 from the right and the left.
