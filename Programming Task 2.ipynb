```python
# write your code here 
def BMI(weight, height):
    'takes values for weight and height and BMI range'
    import math
    if (weight * 703) / (math.pow(height, 2)) < 18.5:
        print("Underweight")
    elif (weight * 703) / (math.pow(height, 2)) <= 25:
        print("Normal")
    else:
        print("Overweight")
```

```python
BMI(190, 75)  # should print Normal
```

```python
# write your code here 
def arithmetic(list):
    'returns True if list is an arithetic sequence, False otherwise'
    if len(list) == 1:
        return True
    var = list[1] - list[0]
    for i in range(len(list)-1):
        if not list[i+1] - list[i] == var:
            return False
    else:
        return True
```

```python
arithmetic([3, 6, 9, 12, 15]) 
```

```python
arithmetic([3, 6, 9, 11, 14]) 
```

```python
arithmetic([3])
```

```python
# write your code here 
def divisors(n):
    'returns a list of all all positive divisors of an integer n'
    lst = []
    for i in range(1, n+1):
        if n % i == 0:
            lst.append(i)
    print(lst)
            
```

```python
divisors(1)
```

```python
divisors(6)
```

```python
divisors(11)
```

```python
# write your code here 
def bubbleSort(lst):
    'sorts a list of numbers using bubble sort'
    for i in range(len(lst)):
        for j in range(0, len(lst) - i - 1):
            if lst[j] > lst[j + 1]:
                (lst[j], lst[j + 1]) = (lst[j + 1], lst[j])
```

```python
lst = [3, 1, 7, 4, 9, 2, 5]
bubbleSort(lst)
lst
```
