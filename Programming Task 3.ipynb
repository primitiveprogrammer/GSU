```python
# write your code here 
def game(r, c):
    import random
    g = [[0 for row in range(r)] for column in range(c)]
    for row in g:
        print(" ".join(str(cell) for cell in row))
    a = random.randrange(1, r)
    b = random.randrange(1, c)
    bomb = (str(a) + ", " + str(b))
    #print(bomb)
    while True:
        print("Find the bomb. \nEnter your selection as Row, Column:")
        guess = input()
        if guess in bomb:
            break
```

```python
# test your code here
game(4,4)
```

```python
# define your function here
def wordcount(string):
    string = string.lower()
    array = string.split()
    d = {}
    for word in array:
        if word.isalpha():
            if word not in d:
                d[word] = 1
            else:
                d[word] += 1
    return d
```

```python
string = "lime duck Duck 33 table lime 1table1"
print(wordcount(string))
```
